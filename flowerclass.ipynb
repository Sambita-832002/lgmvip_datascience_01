```python
#separate features and target 

data = df.values
X = data[:,0:4]
Y = data[:,4]
```

```python
#calculate average of each features for all clases

Y_Data = npy.array([npy.average(X[:,i][Y==j].astype('float32')) for i in range (X.shape[1]) for j in (npy.unique(Y))])
Y_Data_reshaped = Y_Data.reshape(4,3)
Y_Data_reshaped = npy.swapaxes(Y_Data_reshaped, 0,1)
X_axis = npy.arange(len(column)-1)
width=0.25
```

```python

# plot the average

plt.bar(X_axis, Y_Data_reshaped[0], width, label = 'Setosa')
plt.bar(X_axis+width, Y_Data_reshaped[1], width, label = 'Versicolour')
plt.bar(X_axis+width*2, Y_Data_reshaped[2], width, label = 'Verginica')
plt.xticks(X_axis, column[:4])
plt.xlabel("Features")
plt.ylabel("Values in cm.")
plt.legend(bbox_to_anchor=(1.3,1))
plt.show
```

```python
# split the data to train and test dataset

from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2)
print(X_test, Y_test)
```

```python
#SUPPORT VECTOR MACHINE ALGORITHM

from sklearn.svm import SVC

model_svc = SVC()
model_svc.fit(X_train, Y_train)
```

```python

prediction1 = model_svc.predict(X_test)

# calculate the accuracy

from sklearn.metrics import accuracy_score
print(accuracy_score(Y_test, prediction1)*100)
```

```python
#LOGISTIC REGRESSION
from sklearn.linear_model import LogisticRegression
model_LR=LogisticRegression()
model_LR.fit(X_train, Y_train)
```

```python
prediction2 = model_LR.predict(X_test)

#calculate the accuracy

from sklearn.metrics import  accuracy_score
print(accuracy_score(Y_test, prediction2)*100)
```
